```
import cv2
from google.colab.patches import cv2_imshow
```

```
book_page=cv2.imread("/content/bookpage.jpg")
cv2_imshow(book_page)
```

```
# Syntax: cv2.cvtColor(src, code[, dst[, dstCn]])

# Parameters:
# src: It is the image whose color space is to be changed.
# code: It is the color space conversion code.

# dst: It is the output image of the same size and depth as src image. It is an optional parameter.
# dstCn: It is the number of channels in the destination image.

gray_scale=cv2.cvtColor(book_page,cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_scale)
```

```
# Syntax: cv2.threshold(src,thresh,maxVal,type)

# Parameters:
# src:  source image, which should be a grayscale image
# thresh:  threshold value which is used to classify the pixel values.

# maxVal: value to be given if pixel value is more than (sometimes less than) 
#         the threshold value. 
# type: different styles of thresholding

# cv2.THRESH_BINARY
# cv2.THRESH_BINARY_INV
# cv2.THRESH_TRUNC
# cv2.THRESH_TOZERO
# cv2.THRESH_TOZERO_INV


retval, threshold = cv2.threshold(gray_scale, 10, 255, cv2.THRESH_BINARY)
cv2_imshow(threshold)
```

```
# Syntax: cv2.adaptiveThreshold(	src, maxValue, adaptiveMethod, thresholdType, blockSize, C[, dst]	)


# Parameters:

# src :	Source 8-bit single-channel image.
# maxValue : Non-zero value assigned to the pixels for which the condition is satisfied
# adaptiveMethod	: thresholding algorithm to use
# thresholdType :	Thresholding type that must be either THRESH_BINARY or THRESH_BINARY_INV
# blockSize :	Size of a pixel neighborhood that is used to calculate a threshold value for the pixel: 3, 5, 7, and so on.
# C	: Constant subtracted from the mean or weighted mean 


adaptive_threshold = cv2.adaptiveThreshold(gray_scale,255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 215, 1)
cv2_imshow(adaptive_threshold)
```

```
# path to input images are specified and   
# images are loaded with imread command  
image_back = cv2.imread('/content/image_back.png')  
image_circle = cv2.imread('/content/image_circle.png') 
  
cv2_imshow(image_back)
```

```
cv2_imshow(image_circle)
```

```
# cv2.bitwise_and is applied over the 
# image inputs with applied parameters  
image_and = cv2.bitwise_and(image_back, image_circle, mask = None) 
cv2_imshow(image_and)
```

```
image_or = cv2.bitwise_or(image_back, image_circle, mask = None) 
cv2_imshow(image_or)
```

```
image_not_circle = cv2.bitwise_not(image_circle, mask = None)
cv2_imshow(image_not_circle)
```

```

ball_track=cv2.imread("/content/ball-track.png")
test=ball_track.copy()
cv2_imshow(ball_track)
```

```
# Syntax: cv2.GaussianBlur(src, ksize, sigmaX[, dst[, sigmaY[, borderType=BORDER_DEFAULT]]] )

# src   :input image
# ksize :Gaussian Kernel Size. [height width]. height and width should be odd and can have different values. 
#        If ksize is set to [0 0], then ksize is computed from sigma values.

gaussian_blur= cv2.GaussianBlur(ball_track, (11, 11), 0)
cv2_imshow(gaussian_blur)
```

```
hsv = cv2.cvtColor(gaussian_blur, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv)
```

```
# We threshold the HSV image for a range of green color
# Now extract the green object alone, we can do whatever on that image we want.

greenLower = (29, 86, 6)
greenUpper = (64, 255, 255)
mask = cv2.inRange(hsv, greenLower, greenUpper)
cv2_imshow(mask)
```
